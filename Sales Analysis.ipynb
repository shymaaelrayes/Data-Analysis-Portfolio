# Merging Sales Monthly data into one DataFrame

```python
all_months_data = pd.DataFrame()
```

```python
import os
```

```python
files = [file for file in os.listdir('./Sales Data 2019')]
```

```python
files
```

```python
for file in files:
    df = pd.read_csv('./Sales Data 2019/' + file)
    all_months_data = pd.concat([all_months_data, df])
    
```

```python
all_months_data.to_csv('concat_data.csv', index=False)
```

```python
all_data2 = pd.read_csv('concat_data.csv')

all_data2.head()
```

```python
all_data2=all_data2.dropna(how='all')
all_data2.head()
```

```python
all_data2 = all_data2.drop_duplicates()
```

```python
all_data2.dtypes
```

```python
all_data2
```

```python
all_data2['Month'] = all_data2['Order Date'].str[0:2]
all_data2.head()
```

```python
 new_data = all_data2[all_data2['Order Date'].str[0:2] !='Or']
    
```

```python
new_data['Month'] = new_data['Month'].astype('int32')
```

```python
new_data.head()
```

## Second: Exploratory Data Analysis

### Question #1: What was the best month for sales?

```python
new_data['Quantity Ordered'] = new_data['Quantity Ordered'].astype('int32')
new_data['Price Each'] = new_data['Price Each'].astype('float')
new_data['Revenue'] = new_data['Quantity Ordered'] * new_data['Price Each']
```

```python
new_data.head()
```

```python
X = new_data.groupby('Month').sum()
```

```python
X['Revenue'].max()
```

```python
X
```

```python
import matplotlib.pyplot as plt
import calendar
a = new_data['Month'].unique()
a.sort()
aa=[]
for i in a:
      aa.append(calendar.month_name[i])
       
plt.bar(aa, X['Revenue'])
plt.xticks(aa, rotation = 90, size=8)
plt.xlabel('Months')
plt.ylabel('Sales in USD $')
plt.title('Sales per Month')
plt.show()   

```

### Question #2: What city sold the most product?

```python
new_data['city name'] = new_data['Purchase Address'].apply(lambda x: x.split(',')[1] + ' ' + '(' + x.split(',')[2].split(' ')[1] + ')')
new_data.head()
```

```python
b = new_data.groupby('city name').sum()
b
```

```python
c=new_data['city name'].unique()
c.sort()
c
```

```python
plt.bar(c, b['Revenue'])
plt.xticks(c, rotation = 'vertical', size =8)
plt.show()
```

### Question #3: What time should we display advertisements to maximize the likelihood of purchases?

```python
new_data['Order Date'] = pd.to_datetime(new_data['Order Date'])
```

```python
new_data.head()
```

```python
new_data['Hour'] = new_data['Order Date'].dt.hour
```

```python
new_data.head()
```

```python
new_data['Minutes'] = new_data['Order Date'].dt.minute
```

```python
new_data.head()
```

```python
hour = range(0,24)
plt.plot(hour, new_data.groupby('Hour').count())
plt.xticks(hour)
plt.grid()
plt.show()
```

### Question #4: What products are most often sold together?

```python
df = new_data[new_data['Order ID'].duplicated(keep=False)]
df['Bought together'] = df.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))
```

```python
df.head()
```

```python
df = df[['Order ID', 'Bought together']].drop_duplicates()
```

```python
df.head()
```

```python
k=df.groupby('Bought together').count()
```

```python
j=k.sort_values('Order ID', ascending=False)
```

### Question #5: What product sold the most? Why do you think it did? 

```python
new_data.head()
```

```python
v=new_data.groupby('Product').sum()
v
```

```python
s= new_data['Product'].unique()
s.sort()
s
```

```python
plt.bar(s, v['Quantity Ordered'])
plt.xticks(s, rotation=90, size=8)
plt.ylabel('Quantity sold')
plt.title('Quantity sold per product')
plt.show()
```

```python
p1 = new_data.groupby('Product').mean()

fig, ax1 = plt.subplots()
ax2 = ax1.twinx()
ax1.bar(s, v['Quantity Ordered'], color='r')
ax2.plot(s, p1['Price Each'],'b-')

ax1.set_xticklabels(s, rotation ='vertical', size=8)


ax1.set_xlabel('Products')
ax1.set_ylabel('Quantity sold', color = 'g')
ax2.set_ylabel('Price Each', color ='b')

plt.show()
```
